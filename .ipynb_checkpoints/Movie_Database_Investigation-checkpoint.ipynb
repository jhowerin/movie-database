### Data Cleaning (Replace this with more specific notes!)

```python
# The Data Wranling section highlighed the issues with the initial data sets
# In this section, our Data Cleaning will include the following:
# 1) drop duplicates
# 2) delete any rows empty/zero values in a few select columns
# 3) we'll adjust budget_adj and revenue_adj to int64, they do not need to be float64
# 4) We'll adjust release_date to a datetime, it does not need to be a string object
```

```python
# Data Cleaning for the movies dataframe
# drop duplicates
movies.drop_duplicates(keep='first', inplace=True)
sum(movies.duplicated())
```

```python
# Data Cleaning for the movies dataframe
# remove rows with missing (zero's) data in the budget, revenue, budget_adj, revenue_adj columns
flag = ['budget', 'revenue', 'budget_adj', 'revenue_adj', 'vote_count']
movies[flag] = movies[flag].replace(0,np.NaN)
movies.dropna(subset=flag,inplace=True)
movies.info()
```

```python
# adjust budget, revenue, budget_adj and revenue_adj to int64, they do not need to be float64
movies['budget'] = movies['budget'].astype('int64')
movies['revenue'] = movies['revenue'].astype('int64')
movies['budget_adj'] = movies['budget_adj'].astype('int64')
movies['revenue_adj'] = movies['revenue_adj'].astype('int64')
# adjust release_date to a datetime
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies.info()
```

```python
# Get Profit and Adjusted Profit. Add to the file
profit = movies['revenue']-movies['budget']
profit_adj = movies['revenue_adj']-movies['budget_adj']
movies.insert(5,'profit',profit)
movies.insert(22,'profit_adj',profit_adj)
movies['profit'] = movies['profit'].astype('int64')
movies['profit_adj'] = movies['profit_adj'].astype('int64')
# movies['profit'] = movies.apply(lambda x: "{:,}".format(x['profit']), axis=1)
# movies['profit_adj'] = movies.apply(lambda x: "{:,}".format(x['profit_adj']), axis=1)
display(movies)
```

```python
# Function to calculate profit values
def highest_lowest(colName):
    # highest
    highest_id =  movies[colName].idxmax()
    highest_details = pd.DataFrame(movies.loc[highest_id])
    # lowest
    lowest_id = movies[colName].idxmin()
    lowest_details = pd.DataFrame(movies.loc[lowest_id])
    # concat the two DF's
    hiLo = pd.concat([highest_details,lowest_details],axis=1)
    return hiLo

# call the highest_lowest fn
highest_lowest('profit')
```

```python
# call the highest_lowest runtime
highest_lowest('runtime')
```

```python
# call the highest_lowest budget
highest_lowest('budget')
```

```python
# call the highest_lowest revenue
highest_lowest('revenue')
```

```python
# Let's look at averages
def average_func(column_name):
    return movies[column_name].mean()
```

```python
# Average runtime
average_func('runtime')
```

```python
def extract_data(column_name):
    all_data = movieProfit[column_name].str.cat(sep = "|")
    all_data = pd.Series(all_data.split('|'))
    count = all_data.value_counts(ascending = False)
    return count
```

```python
director_count = extract_data('director')
director_count.head()
```

```python
cast_count = extract_data('cast')
cast_count.head()
```

```python
# do profit per cast_count
```

```python
genre_count = extract_data('genres')
genre_count.head()
```

```python
genre_count.sort_values(ascending=True, inplace=True)
ax = genre_count.plot.barh(color='#007482', fontsize = 15)
ax.set(title='The Most Filmed Genres')
ax.set_xlabel("Number of Movies", color='g',fontsize = '18')
ax.figure.set_size_inches(12,10)
plt.show()
```

```python
index_release_date = movieProfit.set_index('release_date')
groupby_index = index_release_date.groupby([(index_release_date.index.month)])
monthly_movie_count = groupby_index.profit.count()
monthly_movie_count = pd.DataFrame(monthly_movie_count)
```

```python
month_list = ['January', 'February', 'March', 'April', 'May', 
              'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
monthly_movie_count_bar = sns.barplot(x=monthly_movie_count.index, y=monthly_movie_count.profit, data=monthly_movie_count)
```

```python
monthly_movie_count_bar = sns.barplot(x=monthly_movie_count.index, y=monthly_movie_count.profit, data=monthly_movie_count)
monthly_movie_count_bar.figure.set_size_inches(15,8)
monthly_movie_count_bar.axes.set_title('Number of Movies Released / Month', color='r', fontsize=25, alpha=0.6)
monthly_movie_count_bar.set_xlabel("Months", color='g',fontsize=25)
monthly_movie_count_bar.set_ylabel("Number of Movies", color='y',fontsize=35)
monthly_movie_count_bar.tick_params(labelsize=15, labelcolor='black')
monthly_movie_count_bar.set_xticklabels(month_list, rotation=30, size = 18)
plt.show()
```

```python
monthly_profit=groupby_index.profit.sum()
monthly_profit = pd.DataFrame(monthly_profit)
monthly_profit_bar = sns.barplot(x=monthly_profit.index, y=monthly_profit.profit, data=monthly_profit)
monthly_profit_bar.figure.set_size_inches(15,8)
monthly_profit_bar.axes.set_title("Profits Made by Movies per Release Month", color='r', fontsize=25, alpha=0.6)
monthly_profit_bar.set_xlabel("Months", color='g', fontsize=25)
monthly_profit_bar.set_ylabel("Profits", color='y', fontsize=35)
monthly_profit_bar.tick_params(labelsize=15, labelcolor='black')
monthly_profit_bar.set_xticklabels(month_list, rotation=30, size = 18)
plt.show()
```
